# Data Visualisation

```python
# Average house prices trend in London since 1995 to 2024

plt.figure(figsize=(12, 6))
plt.plot(data_1['Date'], data_1['LONDON'], marker='o')
#marker = s means square type and line style double dash means dashed line
#plt.plot(data_1['Date'], data_1['LONDON'], marker='s', linestyle='--', color='blue')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.title('Average House Prices in London (1995-2024)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 cities which have highest house prices in London

# Filter the data for the year 2024
data_2024 = data_1[data_1['Date'].dt.year == 2024]

# Calculate the average price for each city in 2024
average_prices = data_2024.iloc[:, 1:].mean()

# Get the top 10 highest average price cities
top_10_cities = average_prices.nlargest(10)

# Plot the bar chart
plt.figure(figsize=(10, 7))
top_10_cities.plot(kind='bar', edgecolor='black')
plt.xlabel('City')
plt.ylabel('Average Price (2024)')
plt.title('Top 10 Highest Average House Prices Cities in 2024')
plt.xticks(rotation=45)
plt.tight_layout()

# Hide Y-axis label and values
plt.yticks([])
plt.ylabel('')

# Show the actual price of the bars on the plot
for i, value in enumerate(top_10_cities):
    plt.text(i, value, f'{value:.0f}', ha='center', va='bottom')

plt.tight_layout()

plt.show()
```
